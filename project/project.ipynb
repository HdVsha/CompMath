```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt

t0tf = np.array([0, 20])  # [t0, tf]
x0 = np.array([2, 0])     # [x1, x2]
N = 1000000                                  # Number of sub intervals
T = np.linspace(t0tf[0], t0tf[1], N+1)
```

```python
def vdp(x, t, a):       # Switched the t and x from before
    ''' Vander Pol equation
    :param t: time
    :type t: int/float
    :param x: state vector, [x1, x2]
    :type x: np.array
    :xpoint: derivative vector
    '''
    x1, x2 = x
    return np.array([x2, -a*((-1 + x1 ** 2) * x2 + x1)])

def run_vanderpol(fun, x, t, a):
   ''' To call the scipy.integrate.odeint '''
   y = scipy.integrate.odeint(fun, x, t, args=(a,))
   return y

Y = run_vanderpol(vdp, x0, T, a=1000)
y1 = []
y2 = []
for i in range(len(Y)):
  y1.append(Y[i][0])
  y2.append(Y[i][1])

fig1 = plt.figure(figsize=(16, 8))
ax1 = fig1.add_subplot(1, 2, 1)
ax1.set_title("График при a = 1000")
ax1.scatter(y1,y2)
Y = run_vanderpol(vdp, x0, T, a=1)
y1 = []
y2 = []
for i in range(len(Y)):
  y1.append(Y[i][0])
  y2.append(Y[i][1])
ax2 = fig1.add_subplot(1, 2, 2)
ax2.set_title("Графики при a = 1")
ax2.scatter(y1,y2)

fig2 = plt.figure(figsize=(16, 8))
ax1 = fig2.add_subplot(1, 2, 1)
ax1.set_title("Графики при a = 0.1")

Y = run_vanderpol(vdp, x0, T, a=0.1)
y1 = []
y2 = []
for i in range(len(Y)):
  y1.append(Y[i][0])
  y2.append(Y[i][1])
ax1.scatter(y1,y2)

ax2 = fig2.add_subplot(1, 2, 2)
ax2.set_title("Графики при a = 0.001")

Y = run_vanderpol(vdp, x0, T, a=0.001)
y1 = []
y2 = []
for i in range(len(Y)):
  y1.append(Y[i][0])
  y2.append(Y[i][1])
ax2.scatter(y1,y2)
```
